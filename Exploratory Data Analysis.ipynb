```python
sns.boxplot(y='No. of Bedrooms', data=df1)
plt.title('Boxplot of No. of Bedrooms')
plt.xlabel('Values')
plt.ylabel('No. of Bedrooms')
plt.show()
```

As we can see, most of the houses has 1 or 2 bedrooms where there are fpour datapoints or houses which has 4,5,6, and 7 bedrooms respectively.

```python
sns.boxplot(y='Area', data=df1)
plt.title('Boxplot of Area')
plt.xlabel('Values')
plt.ylabel('Area')
plt.show()
```

Note: The unit for Area is sqrt

As seen in the above boxplot, the minimum dimension of a house starts above 0 and the 1st quartile is somewhere around 700 to 800 sqft. The median of the Area is 1000 which means that most of the middle value of the Area is seen to be 1000 sqft. The 3rd quartile is around 1200 to 1300. Where as the Maximum value is identified to be around 2000 sqft. There are few houses which are too big in dimension and those data points can be considered as outliers.

```python
df1['No. of Bedrooms'].value_counts().plot(kind='bar', color='Orange')

plt.title('Bar Plot of No. of Bedrooms')
plt.xlabel('No. of Bedrooms')
plt.ylabel('Frequency')

# Enable annotations
for i, value in enumerate(df1['No. of Bedrooms'].value_counts()):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
df1['Swimming Pool'].value_counts().plot(kind='bar', color='Orange')

plt.title('Swimming Pool')
plt.xlabel('Categories')
plt.ylabel('Frequency')

# Enable annotations
for i, value in enumerate(df1['No. of Bedrooms'].value_counts()):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
plt.figure(figsize=(100, 10))  # Adjust the width and height as needed

df1['Location'].value_counts().plot(kind='bar', color='Purple', width=0.8)

plt.title('Location')
plt.xlabel('Locations')
plt.ylabel('Frequency')

# Enable annotations
for i, value in enumerate(df1['Location'].value_counts()):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```

With this chart we can get to know that the top 5 locations of the houses in our dataset are Kharghar, Thane West, Mira Road east, Ulwe and Borivali west and the number of houses in these locations are 533,418,390,319 and 176 respectively.

```python
top_10_locations = df1['Location'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))  # Adjust the width and height as needed

top_10_locations.plot(kind='bar', color='Purple', width=0.8)

plt.title('Top 10 Locations')
plt.xlabel('Locations')
plt.ylabel('Frequency')

# Enable annotations
for i, value in enumerate(top_10_locations):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
top_10_Dimensions = df1['Area'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))  # Adjust the width and height as needed

top_10_Dimensions.plot(kind='bar', color='Red', width=0.8)

plt.title('Top 10 Dimensions')
plt.xlabel('Square Feet')
plt.ylabel('Frequency')

# Enable annotations
for i, value in enumerate(top_10_Dimensions):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```

# Multivariate Analysis

```python
sns.scatterplot(x='Area', y='Price', data=df1)
plt.title('Scatter Plot of Area vs. Price')
plt.show()
```

This Shows that the correlation between Area and Price of the house is obvious

```python
sns.scatterplot(x='No. of Bedrooms', y='Price', data=df1)
plt.title('Scatter Plot of No. of Bedrooms vs. Price')
plt.show()
```

```python
sns.pairplot(df1, vars=['No. of Bedrooms', 'Area', 'Price'])
plt.show()
```

```python
correlation_matrix = df1[['Price', 'Area', 'No. of Bedrooms']].corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

We have the correlation matrix of the complete dataset above

```python
sns.boxplot(x='Area', y='No. of Bedrooms', hue='Gymnasium', data=df1)
plt.title('Box Plot of column2 with column1 and column3')
plt.show()
```

```python
sns.regplot(x='Area', y='Price', data=df1, color='Green')
plt.title('Scatter Plot with Correlation Line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

This clearly shows that there is a obvious correlation between the price and Area attributes of the dataset. We have to make sure that the Area attribute is included in the features while building a model as it has a high impact on the price of the house.
