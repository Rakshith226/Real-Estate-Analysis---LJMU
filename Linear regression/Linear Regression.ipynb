```python
## preparing the Data for Modelling
#As we dont have many categorical values and the amenities are already represented in 1's and 0's, there isn't much data preparation required
# We only have to create Dummy variables for the location parameter.
```

```python
df1['No. of Bedrooms'].unique()
```

We now explore locations which is a categorical data. We need to apply the technique of dimensionality reduction to reduce the number of locations.

```python
df1.Location.value_counts()
```

```python
df1.Location = df1.Location.apply(lambda x : x.strip())
```

```python
Location_stats = df1['Location'].value_counts(ascending=False)
Location_stats
```

```python
Location_stats.values.sum()
```

```python
len(Location_stats[Location_stats<=10])
```

Dimensionality Reduction. Any location having less than 10 data points will be tagged as 'other' locations. This way number of categories can be reduced by huge amount.

```python
Location_stats_less_than_10 = Location_stats[Location_stats<=10]
Location_stats_less_than_10
```

```python
df1.Location = df1.Location.apply(lambda x : 'other' if x in Location_stats_less_than_10 else x)
```

```python
df1.head(10)
```

# Dummy Variables

Creating dummy variables for the Location parameter

```python
dummies = pd.get_dummies(df1.Location)
dummies.head()
```

```python
#Concating the dummy dataframe with the original one
df2 = pd.concat([df1,dummies.drop('other',axis='columns')],axis='columns')
df2.head()
```

```python
#Dropping the Location column as we have already created dummy variables for the column
df2.drop(['Location'],axis='columns',inplace=True)
```

# Dividing into X and Y sets for Model Building

```python
#Dividing into X and Y sets for the model building
y_train = df_train.pop('Price')
X_train = df_train
```

```python
# Importing RFE and LinearRegression
# Using Recursive feature elimination(RFE) for automated Feature Selection approach
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10

lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
estimator = lm
```

```python
rfe = RFE(estimator=estimator, n_features_to_select=20, step=1) # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the linear model
lm = sm.OLS(y_train,X_train_rfe).fit()   
```

```python
lm.summary()
```

```python
X_train_1= X_train_rfe.drop('const',axis=1)
```

```python
#checking the VIF values for the above model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_1
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#Dropping the Kalyan East variable as it has P-value more than 0.05 and is insignificant
X_train1= X_train_rfe.drop('Kalyan East',1)
```

```python
X_train_1= sm.add_constant(X_train1)
lm1 = sm.OLS(y_train,X_train_1).fit() 
```

```python
lm1.summary()
```

```python
X_train_2= X_train_1.drop('const',axis=1)
```

```python
vif = pd.DataFrame()
X = X_train_2
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#Lets drop Sector-18 Ulwe as it has a bit high p-value and see how it impacts the model
X_train2= X_train_1.drop('Sector-18 Ulwe',1)
```

```python
X_train_2= sm.add_constant(X_train2)
lm2 = sm.OLS(y_train,X_train_2).fit() 
```

```python
lm2.summary()
```

```python
X_train_3= X_train_2.drop('const',axis=1)
```

```python
vif = pd.DataFrame()
X = X_train_3
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# Now, all the variables in the model have P-value less than 0.05 and VIF less than 5

## Making the predictions

```python
num_vars=['Area', 'No. of Bedrooms', 'Price']

df_test[num_vars]= scaler.transform(df_test[num_vars])
```

```python
y_test = df_test.pop('Price')
X_test = df_test
```

```python
X_train_2.drop(['const'],axis= 1, inplace=True)
```

```python
# Now let's use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_2.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
```

```python
# Making predictions
y_test_pred = lm2.predict(X_test_new)
```

# Model Evaluation

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_test_pred)
fig.suptitle('Actual vs Predictions', fontsize=20)           
plt.xlabel('Actual', fontsize=18)                          
plt.ylabel('Predictions', fontsize=18)  
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_test_pred)
```
